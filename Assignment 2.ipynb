(5) Write a function to get the scorecard of each match. This function would take the innings value as argument and return two scorecard dataframes each for one team as shown below. So the first dataframe would contain the top 4 scorers of the team who batted first and the top 4 bowlers of the opponent team. And the second dataframe would contain the top 4 scorers of the team who batted second and the top 4 bowlers of the opponent team.

```python
def get_scorecard(data):

    # First innings
    bat1 = data[data['innings']==1].groupby('batter')['runs_off_bat'].sum().sort_values(ascending=False).head(4)
    bowl1 = data[data['innings']==1].groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(4)

    scorecard1 = pd.concat([bat1, bowl1], axis=1)

    # Second innings
    bat2 = data[data['innings']==2].groupby('batter')['runs_off_bat'].sum().sort_values(ascending=False).head(4)
    bowl2 = data[data['innings']==2].groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(4)

    scorecard2 = pd.concat([bat2, bowl2], axis=1)

    return scorecard1, scorecard2
```

```python
for i in range(len(dataset)):

    match_innings_raw = dataset['innings'].iloc[i]

    match_innings_parsed = ast.literal_eval(match_innings_raw)

    rows = []
    for inning_index, inning_detail in enumerate(match_innings_parsed): # Use the parsed data
        for inning_name, inning_content in inning_detail.items():
            team = inning_content['team']
            deliveries = inning_content['deliveries']
            for delivery in deliveries:
                for over_ball, ball_data in delivery.items():
                    batsman = ball_data.get('batsman')
                    bowler = ball_data.get('bowler')
                    # Extract runs scored by batsman from the 'runs' dictionary, default to 0 if not found
                    runs_off_bat = ball_data.get('runs', {}).get('batsman', 0)
                    # Check if a wicket occurred in this delivery, assign 1 if yes, 0 if no
                    is_wicket = 1 if 'wicket' in ball_data else 0
                    rows.append({
                        'innings': inning_index + 1, # 'innings' column represents the inning number (1st, 2nd)
                        'team': team,
                        'batter': batsman,
                        'bowler': bowler,
                        'runs_off_bat': runs_off_bat,
                        'is_wicket': is_wicket
                    })

    match_data_for_scorecard = pd.DataFrame(rows)

    score1, score2 = get_scorecard(match_data_for_scorecard)

    print(f"\n\nMatch : {i+1}\n")
    print("Team 1 Score:\n")
    print(score1)
    print("\nTeam 2 Score:\n")
    print(score2)
```
